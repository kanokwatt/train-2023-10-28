```python
#================================
#+ Data Type Before Casting
#================================
a = True
b = 1
c = 1.5
d = "10"

print("================================")
print("+ Data Type Before Casting")
print("================================")
print(type(a), a)
print(type(b), b)
print(type(c), c)
print(type(d), d)
print("--------------------------------")
print("")

#================================
#+ Data Type After Casting
#================================
a = str(a)
b = bool(b)
c = int(c)
d = float(d)
print("================================")
print("+ Data Type After Casting")
print("================================")
print(type(a), a)
print(type(b), b)
print(type(c), c)
print(type(d), d)
print("--------------------------------")
```

#### การสลับค่าระหว่าง 2 ตัวแปร (Swapping): แบบทั่วไป



```python
x = 10
y = 99

print("==========================================")
print("* Before Swapping: x และ y ")
print("==========================================")

print(f"x = {x}")
print(f"y = {y}")
print("==========================================")
print("* การสลับค่าตัวแปร (Swapping) x กับ y แบบทั่วไป")
print("==========================================")

temp = x
x = y
y = temp

print("temp = x")
print("x = y")
print("y = temp")

print("==========================================")
print("* After Swapping: x และ y ")
print("==========================================")
print(f"x = {x}")
print(f"y = {y}")
```

#### [ข้อควรระวัง] - การอ้างอิงตำแหน่งในหน่วยความจำในตำแหน่งเดียวกัน (Same Memory Address)

ใน Python เมื่อมีการ Assign ค่าตัวแปรต้นฉบับ แบบ Collection ให้กับตัวแปรอื่น จะไม่มีการสร้างตัวแปรใหม่ แต่จะมีการเก็บเพียงแค่ที่อยู่ (Address) ของตัวแปรต้นฉบับ ดังนั้นการเปลี่ยนแปลงใด ๆ ที่เกิดขึ้นกับตัวแปร y ก็จะส่งผลให้เกิดการเปลี่ยนแปลงกับตัวแปร x

```python
print("======================================================================")
print("ระมัดระวังการอ้างอิงตำแหน่งในหน่วยความจำในตำแหน่งเดียวกัน (Same Memory Address)")
print("======================================================================")
a = [1, 2, 3]
print(f"a = {a}") # แสดงสมาชิกใน List ของ a

b = a             # กำหนดให้ b มีตำแหน่งในหน่วยความจำเดียวกับ List ของ a
b.append(4)       # เพิ่มค่าต่อท้าย b
print("b.append(4) # เพิ่มค่าต่อท้าย b")

print(f"b = {b}")
print("----------------------------------------------------------------------")
print("")

print("======================================================================")
print("เราไม่ได้เพิ่มค่าให้กับ a เราเพิ่ม 4 ต่อท้าย b แต่ส่งผลกระทบ a ที่อยู่ในตำแหน่งเดียวกัน")
print("======================================================================")
print(f"a = {a}")
print("----------------------------------------------------------------------")
print("")

print("======================================================================")
print("ตรวจสอบตำแหน่งในหน่วยความจำ (Memory Address)")
print("======================================================================")
print(f"a = {id(a)} # a มีที่อยู่เดียวกับ b")
print(f"b = {id(b)} # b มีที่อยู่เดียวกับ a")
print("----------------------------------------------------------------------")
print("")
```

#### 2. ตัวดำเนินการทางคณิตศาสตร์ (Arithmetic Operator)

```python
x = 10
y = 3

print("================================")
print("+  Arithmetic Operator  ")
print("================================")
print(f"x = {x}") # x
print(f"y = {y}") # y
print("--------------------------------")
print(f"x + y = {x+y}") # add
print(f"x - y = {x-y}") # subtraction
print(f"x * y = {x*y}") # multiplication
print(f"x / y = {x/y}") # division
print(f"x % y = {x%y}") # modulus
print(f"x ** y = {x**y}") # exponential
print(f"x // y = {x//y}") # floor division
print("--------------------------------")
```

#### 3. ตัวดำเนินการที่ใช้ในการเปรียบเทียบ (Comparison Operator)

```python
x = 0.1
y = 0.2
z = 0.3

print("================================")
print("+  Comparison Operator  ")
print("================================")
print(f"x = {x}") # x
print(f"y = {y}") # y
print(f"z = {z}") # z
print("--------------------------------")

print(f"x > y  = {x > y}")
print(f"x < y  = {x < y}")
print(f"x >= y = {x >= y}")
print(f"x <= y = {x <= y}")
print(f"x != y = {x != y}")
print(f"((x + y) == z) = {((x + y) == z)} # ทำไม (0.1 + 0.2) ไม่เท่ากับ 0.3")

print("--------------------------------")
```

#### [วิธีที่การแก้ไขข้อผิดพลาด] (Bug):  0.1 + 0.2 != 0.3 ด้วยฟังก์ชัน round(...)

ข้อผิดพลาดที่เรียกว่า Bug จะแตกต่างจาก Error ที่

#### 5. ตัวดำเนินการด้านเอกลักษณ์ (Identity Operator)

```python
x = ["ant", "bat"]
y = ["ant", "bat"]
z = x
print("=====================================================================================")
print("ระบุให้ตัวแปรแบบ Collection z อ้างอิงตำแหน่งในหน่วยความจำเดียวกัน (Same Memory Address) กับ x ")
print("=====================================================================================")
print(f"x = {x}")
print(f"y = {y}")
print("z = x")
print("-------------------------------------------------------------------------------------")
print("")

print("================================")
print("* Identity Operator")
print("================================")
print(f"x is y = {x is y}")
print(f"x is z = {x is z}")
print(f"y is z = {y is z}")
print("--------------------------------")

print("============================================================")
print("* ตรวจสอบตำแหน่งของการจัดเก็บข้อมูลในหน่วยความจำ (Memory Address)")
print("============================================================")
print(f"address of x is {id(x)}")
print(f"address of y is {id(y)}")
print(f"address of z is {id(z)}")
print("-------------------------------------------------------------")
```

#### [ข้อควรระวัง] - การอ้างอิงตำแหน่งในหน่วยความจำของตัวแปรต้นฉบับ (Reference) ที่เป็น Collection: Sequence (Array-Like)

x = list(...) <br/><br/>

y = x <br/><br/>

เมื่อมีการแก้ไข y จะส่งผลกระทบกับ x ที่อ้างอิงในตำแหน่งของหน่วยความจำเดียวกัน<br/>

```python
print("=====================================================================================")
print("ระบุให้ตัวแปรแบบ Collection z อ้างอิงตำแหน่งในหน่วยความจำเดียวกัน (Same Memory Address) กับ x ")
print("=====================================================================================")

x = [10]
print(f"x = {x}")
z = x
print("z = x   # ทำแบบนี้จะทำให้ z อ้างอิงตำแหน่งของหน่วยความจำเดียวกับ x")
print("-------------------------------------------------------------------------------------")
print("")

z[0] = z[0] + 1
print("z[0] = z[0] + 1   # แก้ไขค่าใน z")
print(f"ค่าใน x = {x}     # แก้ไขค่าใน z แต่ส่งผลต่อค่าใน x")
```

#### **break**
ใช้เพื่อ หยุดการวนซ้ำ เมื่อต้องการ

```python
# แสดงตัวเลขตั้งแต่ 1 - 3
print("======================")
print("* แสดงตัวเลขตั้งแต่ 1 - 3")
print("======================")
for i in range(1, 4):
  print(f"i = {i}")
print("----------------------")
print("")

# หยุดการวนซ้ำเมื่อพบเลข 2
print("======================")
print("* หยุดเมื่อเจอเลข 2")
print("======================")
for i in range(1, 4):
  if i % 2 == 0:
    break
  print(f"i = {i}")
print("----------------------")
```

ใช้เพื่อ ข้ามการวนซ้ำ เมื่อต้องการ

```python
# แสดงตัวเลขตั้งแต่ 1 - 3
print("======================")
print("* แสดงตัวเลขตั้งแต่ 1 - 3")
print("======================")
for i in range(1, 4):
  print(f"i = {i}")
print("----------------------")
print("")

# ข้ามการวนซ้ำเมื่อพบเลข 2
print("======================")
print("* ข้ามการเมื่อพบเลข 2")
print("======================")
for i in range(1, 4):
  if i % 2 == 0:
    continue
  print(f"i = {i}")
print("----------------------")
```

```python
def bmi(weight, height):
  return (weight / ((height/100)**2))  # ค่าที่ส่งกลับ return


print("========================================")
print("* User-defined Function (return 1 value)")
print("========================================")
print("def bmi(weight, height):")
print("  return (weight / ((height/100)**2))")
print("")
print(f"bmi(77, 180) = {round(bmi(77, 180), 2)}  # ค่าที่ส่งกลับ return")
print("----------------------------------------")
```

#####**2. ส่งกลับ (return) มากกว่า 1 ค่า**

```python
def test():
  return 'abc', 99  # ส่งกลับ ชื่อ และ ราคา

print("===========================================")
print("* User-defined Function (return  > 1 value)")
print("===========================================")
print("def test():")
print("  return 'abc', 99  # ส่งกลับ ชื่อ และ ราคา")
print("")
print("name, price = test() ")
print("")
name, price = test()
print("--------------------------------------------")

print(f"name = '{name}'")
print(f"price = {price}")
print("--------------------------------------------")

```

### **4. Dictionary**: A collection which is <i><font color='blue'>ordered</font></i> and <i><font color='blue'>changeable</font></i>. <i><font color='red'> No duplicate</i></font> members.

จัดเก็บข้อมูลในรูปแบบ **key : value** โดยการเข้าถึงข้อมูล **(value)** ผ่าน **key**

```python
car = {
    "brand": "Honda",
    "model": "Jazz",
    "color": "Pink",
    "speed": 180
}

print("=====================================================================================")
print("* Dictionary จัดเก็บข้อมูลในรูปแบบ key : value โดยการเข้าถึงข้อมูล (value) ผ่าน key")
print("=====================================================================================")
print("")
print(f"Dictionary of car = {car}")
print("-------------------------------------------------------------------------------------")
print("")

print("==========================================")
print("* การเข้าถึงข้อมูล (value) ผ่าน key")
print("==========================================")
print(f"car['brand'] = {car['brand']}")
print(f"car['model'] = {car['model']}")
print(f"car['color'] = {car['color']}")
print(f"car['speed'] = {car['speed']}")
print("------------------------------------------")
print("")

print("==========================================")
print("* การเข้าถึงข้อมูล (key, value) ผ่าน .items()")
print("==========================================")
print("for key, value in car.items():")
print('  print(f"{key}: {value}")')
print("")

for key, value in car.items():
    print(f"{key}: {value}")

print("")
print("------------------------------------------")

print("==========================================")
print("* การเข้าถึงข้อมูล (key) ผ่าน .keys()")
print("==========================================")
print("for key in car.keys():")
print('  print(f"{key}")')
print("")

for key in car.keys():
    print(f"{key}")

print("")
print("------------------------------------------")

print("==========================================")
print("* การเข้าถึงข้อมูล (value) ผ่าน .values()")
print("==========================================")
print("for value in car.values():")
print('  print(f"{value}")')
print("")

for value in car.values():
    print(f"{value}")

print("")
print("------------------------------------------")
```

## **ต้นแบบ (Class)**

### การสร้างต้นแบบ (Class) ของรถยนต์

```python
# ต้นแบบรถยนต์ประกอบด้วยคุณสมบัติ (Property) และพฤติกรรม (Method / Function)
class Car:
    def __init__(self, brand, speed, color): # คุณสมบัติ (Property)
        self.brand = brand  # ยี่ห้อรถยนต์
        self.speed = speed  # ความเร็ว
        self.color = color  # สี

    def detail(self):       # พฤติกรรม (Method / Function) ของรถ
        print(f"{self.brand}: {self.color}")
        print(f"Speed: {self.speed}")
```

```python
car1 = Car("Toyota", 180, "White") # การสร้างวัตถุที่เป็นรถยนต์ มีชื่อว่า car1
```

```python
car1.detail() # เรียกใช้งาน car1 เพื่อให้แสดงรายละเอียดของรถ
```

### การถ่ายทอด(Inheritance) ต้นแบบของรถยนต์ -> รถไฟฟ้า

```python
class Ecar(Car):  # การถ่ายทอด(Inheritance) ต้นแบบของรถยนต์ -> รถไฟฟ้า
    def __init__(self, brand, speed, color, energy):
        super().__init__(brand, speed, color) # super() คือใช้ของต้นแบบ
        self.energy = energy                  # เพิ่มคุณสมบัติ (energy)

    def fast_charge(self, x):                 # พฤติกรรม (Method / Function) ของรถไฟฟ้า
        print("Car energy is " + str(self.energy) + " %, and charging at " + str(x) + " KW" )
        self.energy = x
```

```python
car2 = Ecar("Tesla", 220, "Black", 10) # การสร้างวัตถุที่เป็นรถยนต์ไฟฟ้า มีชื่อว่า car2
```

```python
car2.detail() # เรียกใช้งาน car2 เพื่อให้แสดงรายละเอียดของรถ
```

```python
car2.fast_charge(150) # เรียกใช้งาน car2 เพื่อให้แสดงรายละเอียดของพลังงาน และกำลังของการประจุพลังงานไฟฟ้า
```
